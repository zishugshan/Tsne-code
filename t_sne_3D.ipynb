```python
import numpy
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
# create multidimensional data

A = numpy.random.normal(2.0, scale=0.5, size=(40,3)) 
B = numpy.random.normal(-2.0, scale=0.5, size=(40,3)) 
C = numpy.random.normal(0.0, scale=0.5, size=(40,3)) 
D = numpy.random.normal(0.0, scale=0.5, size=(40,3)) 
D[:,2] = D[:,2] + 3
E = numpy.random.normal(0.0, scale=0.5, size=(40,3)) 
E[:,0] = E[:,0] + 3
F = numpy.random.normal(0.0, scale=0.5, size=(40,3)) 
F[:,0] = F[:,0] - 2
G = numpy.random.normal(0.0, scale=0.5, size=(40,3)) 
G[:,1] = G[:,1] + 2
data = numpy.concatenate((A, B, C, D, E, F, G), axis=0)

# data should be 280 3-dimensional points
data.shape
```

```python
# visualise 2d slices of 3d data

fig = plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(131)
ax1.scatter(data[:,0], data[:,1])
ax2 = fig.add_subplot(132)
ax2.scatter(data[:,0], data[:,2])
ax3 = fig.add_subplot(133)
ax3.scatter(data[:,1], data[:,2])
plt.show()
```

```python
# visualise multi-dimensional data

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2], c='r', marker='o')
plt.show()
```

```python
# create tsne object which will reduce data to 2 dimensions

tsne = TSNE(n_components=2, perplexity=20)
```

```python
# apply tSNE to data

data_2d = tsne.fit_transform(data)
```

```python
# plot tSNE transformed data

plt.scatter(data_2d[:,0], data_2d[:,1])
```
